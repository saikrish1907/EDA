```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#read data from the file using read_csv command
data = pd.read_csv('COVID clinical trials.csv')

#displays first few rows 
data.head()
```

```python
#displays the total rows and columns in the data set 
data.shape
```

```python
#prints the data type of the column
data['Rank'].dtype
```

```python
#returns True for any null value or False if values exists
data.isnull()
```

```python
#now te be more specific check for missing values using sum() and mean() help to fetch the 
#% of missing values in each column.

data.isnull().mean().sort_values(ascending = False)

#here we can see that Result First Posted column has the most null values
#least is Other ID's 
```

```python
#once checked we can either drop the null value using dropna() or you can fill it with fillna()
for col in data:
    if type(data[col]) == 'object':
        data[col] = data[col].fillna(value='UNKNOWN')
    else :
        data[col] = data[col].fillna(value=0)
```

```python
#rechecking for missing values 
data.isnull().mean().sort_values(ascending = False)

#No missing values : Great we are good to go for our data plotting and graph
```
